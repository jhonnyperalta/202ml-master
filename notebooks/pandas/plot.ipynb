We start by setting up the necessary import statements.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
```

First we load the CSV data. For convenience I renamed the CSV file and stripped out the first few rows.

This is essential, because the pandas read_csv function will interpret the first row of the CSV file as a list of column headings.

```python
all_gdp = pd.read_csv('GDP.csv')
```

One peculiarity of read_csv is that it does not interpret the first column as an index column. Instead, the DataFrame that read_csv returns has a plain vanilla index which consists of a sequence of integers: 0,1,2,....

The fix for this problem is to use the set_index method on the DataFrame, telling it to use the 'Country Code' column as the index for the DataFrame. This also removes the first column from the DataFrame.

```python
gdp = all_gdp.set_index('Country Code')
```

We won't need all of the columns of this DataFrame. We just need columns for the years 1995 and 2005. The following notation picks out just those two columns.

```python
gdp = gdp[['1995','2005']]
```

If we wish, we can take a look at our DataFrame now to see that it looks right. Note the NaN values caused by missing data.

```python
gdp
```

Now we read the JSON data and convert it to a DataFrame. To do this, I read the JSON file the same we did in an earlier example. Next, I construct an empty DataFrame to receive the data values. We then loop over the JSON data looking for dictionaries with the correct values for the year. When we come across and object that refers to 1995 or 2005 we copy a population value into the DataFrame.

The syntax we use to assign new values to the DataFrame is

pop.at[<country code>,<year>] = <value>

The .at[] construct is what pandas uses to assign values to the cells of a DataFrame.

```python
# Read the raw JSON data
with open('population_data.json') as f:
    pop_data = json.load(f)

# Prepare a DataFrame to receive the population data
pop = pd.DataFrame(np.nan,index=gdp.index,columns=['1995','2005'])
# Process the population data, selecting only those entries for
# the years 1995 and 2005
for p in pop_data:
    if p['Year'] == '1995' or p['Year'] == '2005':
        pop.at[p['Country Code'],p['Year']] = float(p['Value'])
```

Let's take a quick look at the pop DataFrame.

```python
pop
```

Note that this has four more rows than the gdp DataFrame. This is because the population data has entries for two obscure countries not covered by the gdp data set and also adds two more aggregate data points for high income OECD countries (OEC), and everyone else (NOC).

Note also that once again we have some NaNs due to missing data.

Now the computations: we make use some vector calculations involving the data Series in our two DataFrames.

```python
gdp_pc_1995 = gdp['1995']/pop['1995']
gdp_pc_2005 = gdp['2005']/pop['2005']
gdp_growth = (gdp_pc_2005-gdp_pc_1995)/gdp_pc_1995
pop_growth = (pop['2005']-pop['1995'])/pop['1995']
```

The next step is to reassemble these Series into a DataFrame, drop rows containing NaN values, and then plot.

The handy pandas dropna() method drops rows that have any NaNs.

```python
combined = pd.DataFrame({'p':pop_growth,'g':gdp_growth})
filtered = combined.dropna()
plt.plot(filtered['p'],filtered['g'],'bo')
plt.show()
```

This is the plot we are looking for, but it is negatively impacted by one data point with an immensely large gdp growth. This outlier messes up the rest of the graph.

Let's see who that outlier is.

```python
filtered[filtered['g'] > 40]
```

The culprit here is GNQ, Equatorial Guinea. In 1995 this was a small, poor country in west Africa. Sometime after 1995 an immense offshore oil field opened up outside the country, bringing in a torrent of money.

Let's redo the plot with that outlier filtered out.

```python
final = filtered[filtered['g'] < 40]
plt.plot(final['p'],final['g'],'bo')
plt.show()
```

Much better: now we can see what is going on here. Eyeballing this data briefly we can conjecture that there is a mild negative correlation between population growth and gdp growth.

One criticism of this plot is that it puts every country on an equal footing, with one dot per country. Since countries differ quite a bit in their size and gdp levels, we could make this plot more informative by mixing in population and gdp data.

Let's start by making a new DataFrame that features more useable data.

```python
combined = pd.DataFrame({'pg':pop_growth,'gg':gdp_growth,'w':gdp_pc_2005,'p':pop['2005']})
combined = combined[combined['gg']<40]
final = combined.dropna()
```

Now we apply some pyplot magic. We switch from using plot() to using scatter(), since scatter() includes a slick option to set the size of the dots from a data series that we provide. In addition, we apply some pyplot color magic to link the dot colors to the GDP per capita figure. Countries with a low GDP per capita will be drawn in 'cool' colors (purple, then blue), while high GDP countries get drawn in 'hot' colors (orange and red).

```python
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=np.log2(100), vmax=np.log2(5.5e4))
plt.scatter(final['pg'],final['gg'],s=(np.log2(final['p'])**2)/16,color=cmap(norm(np.log2(final['w']))))
plt.show()
```

This plot seems to show a tale of two sets of countries. Poor countries tend to have a higher population growth rate overall, with about the same variability in GDP growth rates as richer countries. Richer countries look a lot like poorer countries, except for the fact that their population growth rates a lower overall.

There are some interesting outliers here. Let's use pandas handy filtering options to look at two sets of outliers, countries with high GDP growth rates and countries with particularly high population growth rates.

```python
final[final['gg']>2]
```

```python
final[final['pg']>0.6]
```

Many of the high GDP growth rate countries emerged out of the Soviet Union. This lead to a somewhat strange combination of almost no population growth coupled with robust GDP growth.

Other examples in the high GDP growth group are countries like Angola and Bosnia-Herzogovina, which were emerging from civil wars in this time period.

The three highest population growth rate countries include Argentina (which boomed early on as it emerged from a military dictatorship, but then fell into a deep financial crisis), Rwanda (which saw many people flowing back in from neighboring countries after the Rwandan genocide ended), and Qatar (which has been experiencing a long-running GDP and population boom fueled by oil wealth).
